```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/titanic_dataset.csv',index_col="Name")
df.head()
```

```python
sns.relplot(data=df,x="Age",y="Fare",hue="Sex",kind="line",col="Pclass")
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()

```

```python
sns.boxplot(x='Pclass', y='Fare',data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
sns.scatterplot(x="Age", y="Fare", data=df, hue="Survived")
plt.title("Age vs. Fare with Survival")
```

```python
sns.scatterplot(x="Age", y="Fare", data=df, hue="Sex")
plt.title("Age vs. Fare with Sex")
```
