```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://jobs.nike.com/?jobSearch=true&jsOffset=0&jsSort=posting_start_date&jsLanguage=en&error=jobPost")
page_data = loader.load().pop().page_content
```

```python
from langchain_core.prompts import PromptTemplate

# Define the prompt template for extracting job postings
prompt_extract = PromptTemplate.from_template(
    """
     ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    The scraped text is from the career's page of a website.
    Your job is to extract the job postings and return them in JSON format containing the following keys: `role`, `experience`, `skills`, and `description`.
    Only return the valid JSON.
    ### VALID JSON (NO PREAMBLE):
    """
)

# Assuming `llm` is defined and set up to invoke the prompt
chain_extract = prompt_extract | llm
res=chain_extract.invoke(input={'page_data': page_data})
print(res.content)
  # Add this line to inspect the loaded page content
```

```python
from langchain_core.output_parsers import JsonOutputParser
json_parsor = JsonOutputParser()
json_parsor.parse(res.content)
```
